```python
import pandas as pd

airbnb = pd.read_csv("listingsdec21.csv")

new_bnb2 = airbnb.copy()

list(airbnb.columns)

new_bnb2 = airbnb[["name","host_location", "latitude", "longitude", "bedrooms", "price"]]
```

```python
new_bnb2
```

```python
# remove o símbolo $ da coluna "valores"
new_bnb2['price'] = new_bnb2['price'].replace({'\$': ''}, regex=True)
new_bnb2

```

```python
new_bnb2['price'] = new_bnb2['price'].replace('[\,]', '', regex=True).astype(float)
new_bnb2
```

```python
new_bnb2['total_price'] = new_bnb2['price'].apply(lambda x: x * 30)
new_bnb2
```

```python
new_bnb2[['Sub_City', 'City', 'Country']] = new_bnb2['host_location'].str.split(',', n=2, expand=True)
new_bnb2
```

```python
new_order = ['name', 'Country', 'City', 'Sub_City', 'bedrooms', 'price', 'total_price', 'latitude','longitude']
```

```python
new_bnb3 = new_bnb2.reindex(columns=new_order)
```

```python
new_bnb3
```

```python
dificil = new_bnb3[['Sub_City', 'City', 'Country']].apply(lambda x: x.str.contains('Portugal|PT')).any(axis=1)
```

```python
new_bnb4 = new_bnb3.loc[dificil]
```
